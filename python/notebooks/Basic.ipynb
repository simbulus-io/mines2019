# Plotting with MatPlotLib (and Pandas)
* Example of using pandas to define a DataFrame
* Plotting to matplotlib
* Note the %matplotlib inline directive
* The None at the end of the code suppresses and Ouput [] block from appearing (try removing "None" and running)
* Try replacing kind="line" with kind="box" and rerunning

```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd

df = pd.DataFrame({
     "2018":[1.3, 1.1, 3.2, 3.7], 
     "2019":[2.1, 3.1, 4.5, 1.2]},
     index=['Jan','Feb','Mar', 'Apr'])

df.plot(kind='box',figsize=(15,10),fontsize=14,grid=False)
plt.ylabel('Rainfall')
None
```
